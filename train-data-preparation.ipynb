```python
import numpy as np
import os
import zipfile
import random
from shutil import copyfile
import re
```

Directory

```python
BASE_DIR = "/kaggle/input/dogs-vs-cats"
PICS_DIR = "/kaggle/working/"
```

Unzip and extract

```python
# Unzip the archive
local_zip = os.path.join(BASE_DIR, "train.zip")
zip_ref = zipfile.ZipFile(local_zip, 'r')
zip_ref.extractall()
zip_ref.close()
```

Splitting the data

```python
def split_data(SOURCE_DIR, TRAINING_DIR, VALIDATION_DIR, SPLIT_SIZE):
  """
  Splits the data into train and test sets
  
  Args:
    SOURCE_DIR (string): directory path containing the images
    TRAINING_DIR (string): directory path to be used for training
    VALIDATION_DIR (string): directory path to be used for validation
    SPLIT_SIZE (float): proportion of the dataset to be used for training
    
  Returns:
    None
  """
  shuffled_source = random.sample(os.listdir(SOURCE_DIR), len(os.listdir(SOURCE_DIR)))

  # Find total number of files in training dir
  training_number = int(len(shuffled_source) * SPLIT_SIZE)

  i = 0
  target = TRAINING_DIR

  for item in shuffled_source:
    item_source = os.path.join(SOURCE_DIR, item)
    if os.path.getsize(item_source) == 0:
      print(f'{item} is zero length, so ignoring.')
    else: 
      copyfile(item_source, os.path.join(target, item))
      i += 1

    # Switch copy target to TESTING
    if i == training_number:
      target = VALIDATION_DIR
```

```python
os.makedirs(os.path.join(f'{PICS_DIR}/training', 'dogs'))
os.makedirs(os.path.join(f'{PICS_DIR}/training', 'cats'))
os.makedirs(os.path.join(PICS_DIR, 'validation'))
os.makedirs(os.path.join(f'{PICS_DIR}/validation', 'dogs'))
os.makedirs(os.path.join(f'{PICS_DIR}/validation', 'cats'))
```

```python
TRAINING_CATS_DIR = os.path.join(f'{PICS_DIR}/training', "cats/")
VALIDATION_CATS_DIR = os.path.join(f'{PICS_DIR}/validation', "cats/")

TRAINING_DOGS_DIR = os.path.join(f'{PICS_DIR}/training', "dogs/")
VALIDATION_DOGS_DIR = os.path.join(f'{PICS_DIR}/validation', "dogs/")
SOURCE = os.path.join(PICS_DIR, "train")
split_size = 0.9
```

```python
split_data(SOURCE, TRAINING_CATS_DIR, VALIDATION_CATS_DIR, split_size)
split_data(SOURCE, TRAINING_DOGS_DIR, VALIDATION_DOGS_DIR, split_size)
```
